The metrics have been designed in the following way (Note that all metrics lie between 0 and 1):
1. ```confidence_inc``` means the labels stay the same but the confidence scores increase (higher the better)
2. ```confidence_dec``` means the labels stay the same but the confidence scores decrease (lower the better)
3. There are separate labels for the two to separately diagnose changes in confidence scores
4. ```iou_arr``` represents the closeness or intersection-over-union for the bounding boxes predictions, closer it is to one means more closeness to the predicted bounding box
5. ```appearance``` and ```disappearance``` are better when lower and higher respectively, they indicate either a loss of prediction or a new prediction. I am operating with the assumption that when new objects are detected, they signify an improvement with greater objects being detected. So, keeping the appearance score low would mean that having the text labels does not impede with the detection. On the contrary, having high disappearance might indicate ...
6. Hypothesis was a high appearance score for crop as cropping might lead to removal of objects from this image. Hypothesis confirmed. Unexpected result was the disappearance being high. It is however also seen that the iou_arr parameter is quite low in cropping so we can conclude there is not a good fit in the bounding boxes of the images. Yet this may be explained by the case that when images were cropped out of their text, no objects were any longer detected in some files. This happened with 80 files in ns and crop, and with 40 files in telea.
7. It is significant to consider these because they show the inefficacy of the alternative methods
8. In conclusion, there is a considerable decrease in perfomance of object detection models when images are cropped or inpainted, probably due to information being distorted. Object detection models perform well enough on text overlaid images, and it is best not to use alternate strategies such as image inpainting or cropping.
9. Yet, if we are to compare, ns and telea inpaintings offer similar performances.
