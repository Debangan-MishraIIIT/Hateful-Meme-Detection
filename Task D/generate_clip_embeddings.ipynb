```python
import pandas as pd
import torch
from PIL import Image
from transformers import AutoTokenizer, AutoProcessor, CLIPTextModelWithProjection, CLIPVisionModelWithProjection
from transformers import CLIPTextConfig
```

```python
model_text = CLIPTextModelWithProjection.from_pretrained("openai/clip-vit-base-patch32")
tokenizer_text = AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")
model_img = CLIPVisionModelWithProjection.from_pretrained("openai/clip-vit-base-patch32")
processor_img = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
with open("./a.txt", "r") as f:
    text= f.read()

temparr=[]
for arrval in text.split("\n"):
    if(arrval==""):
        continue
    arr= [float(val) for val in arrval.split("[")[1][:-1].split(",") if val!=""] 
    temparr.append(arr)

caption_df["text encoding"]=temparr


with open("./b.txt", "r") as f:
    text2= f.read()

temparr2=[]
for arrval in text2.split("\n"):
    if(arrval==""):
        continue
    arr= [float(val) for val in arrval.split("[")[1][:-1].split(",") if val!=""] 
    temparr2.append(arr)

caption_df["cleaned encoding"]=temparr2
```

```python
with open("./img.txt", "r") as f:
    text3= f.read()

temparr3=[]
for arrval in text3.split("\n"):
    if(arrval==""):
        continue
    arr= [float(val) for val in arrval.split("[")[1][:-1].split(",") if val!=""] 
    temparr3.append(arr)

caption_df["image encoding"]=temparr3
```

```python
# Uncomment if needed, used for generating outputs of clip model in text embeddings
with open("./b.txt", "a") as f:
    for i in range(len(caption_df)):
        if(i%400==0):
            print(i)
        txt= caption_df["preprocessed text"][i]
        try:
            inputs = tokenizer_text([txt], padding=True, return_tensors="pt")
            outputs = model_text(**inputs)
            arr= [val.item() for val in outputs.text_embeds[0]]
            f.write(f"{arr}\n")
        except:
            print(i, txt)
            f.write(f"[]\n")

with open("./b.txt", "r") as f:
    text= f.read()

temparr=[]
for arrval in text.split("\n"):
    if(arrval==""):
        continue
    arr= [float(val) for val in arrval.split("[")[1][:-1].split(",") if val!=""] 
    temparr.append(arr)

caption_df["text encoding"]=temparr
with open("./b.txt", "r") as f:
    text= f.read()

temparr2=[]
for arrval in text.split("\n"):
    if(arrval==""):
        continue
    arr= [float(val) for val in arrval.split("[")[1][:-1].split(",") if val!=""] 
    temparr2.append(arr)

caption_df["cleaned encoding"]=temparr2
df_val = caption_df[[len(val)==512 for val in caption_df["text encoding"]]]
pd.to_pickle(df_val, "./df_main2.pkl")
```
