```python
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import os
import pandas as pd
```

```python
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large").to("cuda")
```

```python
count=0
with open("../outputs/captions.txt", "a") as f:
    for path in os.listdir("../hateful_memes/img"):
        if(count%100==0):
            print(count)
        count+=1
        try:
            raw_image = Image.open(f"../hateful_memes/img/{path}")
            text = "a photo without text of a"
            inputs = processor(raw_image, text, return_tensors="pt").to("cuda")
            out = model.generate(**inputs)
            f.write(path+", "+(processor.decode(out[0], skip_special_tokens=True)).replace(text, "")+"\n")
        except Exception as e:
            print(path, e)
```

```python
with open("../outputs/captions.txt") as f:
    text= f.read()
text
```

```python
main_df=[]
for val in text.split("\n"):
    dict={}
    dict["file"]=val.split(",")[0].strip()
    dict["description"]=val.split(",")[1].strip()
    main_df.append(dict)
```

```python
df= pd.DataFrame(main_df)
df
```

```python
df1= pd.read_json("../../hateful_memes/train.jsonl", lines=True)
df1["set"]= "train"
df1
```

```python
df2= pd.read_json("../../hateful_memes/test_seen.jsonl", lines=True)
df2["set"]= "test_seen"
df2
```

```python
df3= pd.read_json("../../hateful_memes/test_unseen.jsonl", lines=True)
df3["set"]= "test_unseen"
df3
```

```python
df4= pd.read_json("../../hateful_memes/dev_seen.jsonl", lines=True)
df4["set"]= "dev_seen"
df4
```

```python
df5= pd.read_json("../../hateful_memes/dev_unseen.jsonl", lines=True)
df5["set"]= "dev_unseen"
df5
```

```python
tempdf= pd.concat((df1, df2, df3, df4, df5))
(tempdf["img"].value_counts()>1).sum()
#this is denoting the number of points which are repeated ion multiple datasets
```

```python
#we see that some values are repeated in two or more sets, so for preprocessing, we will choose to retain both
#idea behind retaining both sets is convenience later on when we need to separate out the datapoints by the set they belong to
#we will also preprocess the file names which have an extra "img/" prefix
#we also have no need for both id and filename
tempdf[tempdf["img"]=="img/32981.png"]
```

```python
tempdf["file_name"]= tempdf["img"].apply(lambda x: x.replace("img/", ""))
tempdf
```

```python
#precaution checking for null values, we are safe here
tempdf.info()
```

```python
#now we will try and merge the two dataframes
result = pd.merge(tempdf, df, left_on='file_name', right_on='file', how='left')
result
```

```python
#we can succesfully see merge is happening correctly
#the captions generated are being duplicated for both tuples
#we will continue with further processing
result[result["img"]=="img/32981.png"]
```

```python
#we again see that the number of hateful/offensive memes are much lesser than normal memes
#this might pose issues with data imbalance later on
#also to mitigate the effects of duplcations in filename, we will need to 
temp_df1= result.drop(["id", "img", "file_name"], axis=1)
temp_df1["label"].value_counts()
```

```python
#we can also choose to store the sets as an array, this gives a proper grouping and allows convenient operations in text processing
result = temp_df1.groupby(['label', 'text', 'file', 'description'])['set'].apply(list).reset_index()
final_result= result.rename(columns={'set': 'sets'}, inplace=False)
final_result
```

```python
final_result.to_pickle("../outputs/caption_description.pkl")
```
