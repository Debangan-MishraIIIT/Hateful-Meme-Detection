```python
from PIL import Image
import pandas as pd
from transformers import CLIPProcessor, CLIPModel, CLIPConfig
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
filename="79846.png"
image = Image.open(f"../../hateful_memes/img/{filename}")
# text1= df.iloc[ind].text
# text2= df.iloc[ind].description
text1= df[df["file"]==filename]["text"].item()
text2= df[df["file"]==filename]["description"].item()

inputs = processor(text=[text1[:77], text2[:77]], images=image, return_tensors="pt", padding=True, max_length=85)
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)  
with open("../outputs/it_similarity.txt", "a") as f:
    f.write(f"{filename}, {logits_per_image[0][0].item()}, {logits_per_image[0][1].item()}\n")
probs
```
